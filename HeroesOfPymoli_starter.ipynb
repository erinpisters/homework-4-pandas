```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#count the number of enteries in SN column. unique returns unique player names. length determines how many
NumPlayers = len(purchase_data["SN"].unique())
NumPlayers

#make this into a dataframe 
NumPlayersdf = pd.DataFrame({"Total Players": [NumPlayers]})
NumPlayersdf.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Find number of unique items
UniqueItems = len(purchase_data["Item ID"].value_counts())


#Find avg price
AvgPrice = purchase_data["Price"].mean()


#Find total revenue 
TotalRev = purchase_data["Price"].sum()


#Find total number of purchases
NumPurchase = (purchase_data["Purchase ID"]).count()


PurchAdf = pd.DataFrame({"Num Items": [UniqueItems], "Average Price": [AvgPrice], 
                                  "Number of Purchases": [NumPurchase],
                                   "Total Revenue": [TotalRev]})
PurchAdf

#Clean Up Data Frame
Sum_PurchAdf = PurchAdf
Sum_PurchAdf["Average Price"] = Sum_PurchAdf["Average Price"].map("${:.2f}".format)
Sum_PurchAdf["Total Revenue"] = Sum_PurchAdf["Total Revenue"].map("${:.2f}".format)
Sum_PurchAdf
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#get rid of players that appear more than once. Tell if duplicate by screen name! 
Demographics = purchase_data.drop_duplicates("SN")
GenderInfo = Demographics["Gender"].value_counts()
GenderPercents = GenderInfo / NumPlayers *100

#Put info into data frame 
GenderInfodf = pd.DataFrame({"Total Count": GenderInfo, "Percentage of Players": GenderPercents})
GenderInfodf.head()

#Clean up Data Frame 
Sum_GenderInfodf = GenderInfodf
Sum_GenderInfodf["Percentage of Players"] = Sum_GenderInfodf["Percentage of Players"].map("{:.2f}%".format)
Sum_GenderInfodf
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#create new data frame that only contains purchase and gender info 
GPdf = purchase_data[["Gender", "Price"]]

#Calc Purchase Count
GPurchaseC = GPdf.groupby(["Gender"]).count()["Price"]

#Calc Avg Purchase Price per Gender
GAvgPurchase = GPdf.groupby(["Gender"]).mean()["Price"]
GAvgPurchase

#Calc Total purchase price by gender
GTotPurchase = GPdf.groupby(["Gender"]).sum()["Price"]
GTotPurchase

#Create a data frame that holds these values! 
gender_panalysis = pd.DataFrame({"Purchase Count": GPurchaseC, "Average Purchase Price": GAvgPurchase,
                                "Total Purchase Value": GTotPurchase})
gender_panalysis.head()

#Clean up Data Frame 
Sum_gender_panalysis = gender_panalysis
Sum_gender_panalysis["Average Purchase Price"] = Sum_gender_panalysis["Average Purchase Price"].map("${:.2f}".format)
Sum_gender_panalysis["Total Purchase Value"] = Sum_gender_panalysis["Total Purchase Value"].map("${:.2f}".format)

Sum_gender_panalysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Construct age bins
age_bin = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 10000000]
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

agedf = purchase_data[["Age", "Price"]]
#pd.cut function to bin the players
agedf["Age Groups"] = pd.cut(agedf["Age"], age_bin, labels = bin_names)
agedf.head()

```

```python
age_totals = agedf["Age Groups"].value_counts()
age_percents = age_totals / NumPlayers *100 

#create the age data frame and sort 
age_summary = pd.DataFrame({"Total count": age_totals, "Percent of Players": age_percents})
age_summary = age_summary.sort_index()

age_summary

#Clean up Data Frame 
Sum_age_summary = age_summary
Sum_age_summary["Percent of Players"] = Sum_age_summary["Percent of Players"].map("{:.2f}%".format)
Sum_age_summary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Cut the data using pd.cut 
age_purchasedf = purchase_data[["Age", "Price"]]
age_purchasedf["Age Groups"] = pd.cut(age_purchasedf["Age"], age_bin, labels = bin_names)
age_purchasedf.head()
```

## Top Spenders

```python
#Perform Calculations 
agep_total = age_purchasedf.groupby(["Age Groups"]).count()["Price"]


age_avgprice = age_purchasedf.groupby(["Age Groups"]).mean()["Price"]

tot_purchase = age_purchasedf.groupby(["Age Groups"]).sum()["Price"]


#create the age data frame and sort 
age_a_summary = pd.DataFrame({"Purchase Count": agep_total, "Average Purchase Price": age_avgprice,
                             "Total Purchase Value": tot_purchase})
age_a_summary = age_a_summary.sort_index()

age_a_summary

#Clean up data frame 
Sum_age_analysis = age_a_summary
Sum_age_analysis["Average Purchase Price"] = Sum_age_analysis["Average Purchase Price"].map("${:.2f}".format) 
Sum_age_analysis["Total Purchase Value"] = Sum_age_analysis["Total Purchase Value"].map("${:.2f}".format)
Sum_age_analysis
```

```python
#Create a data frame with SN and Price 
SN_df = purchase_data[["SN", "Price"]]

#Group by SN and then perform calculations
#Calc Purchase Count
SN_purchasec = SN_df.groupby(["SN"]).count()["Price"]


SN_avg_price = SN_df.groupby(["SN"]).mean()["Price"]



SN_totpurchase = SN_df.groupby(["SN"]).sum()["Price"]


#Create a data frame that holds these values! 
SN_panalysis = pd.DataFrame({"Purchase Count": SN_purchasec, "Average Purchase Price": SN_avg_price,
                                "Total Purchase Value": SN_totpurchase})
SN_panalysis.head()

#Clean up data frame 
Sum_SN_pa = SN_panalysis
Sum_SN_pa["Average Purchase Price"] = Sum_SN_pa["Average Purchase Price"].map("${:.2f}".format)
Sum_SN_pa["Total Purchase Value"] = Sum_SN_pa["Total Purchase Value"].map("${:.2f}".format)
Sum_SN_pa.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Create new data frame with Item ID, Item Name, and Price Columns 
mostpop_df = purchase_data[["Item ID", "Item Name", "Price"]]

#Group by Item ID and then Item Name
g_mostpop = mostpop_df.groupby(["Item ID", "Item Name"])

#Perform Calculations
pop_purchasec = mostpop_df.groupby(["Item ID", "Item Name"]).count()["Price"]

pop_totpurchase = mostpop_df.groupby(["Item ID", "Item Name"]).sum()["Price"]

pop_prices = pop_totpurchase/pop_purchasec

#Create a summary table data frame 
pop_summary_df = pd.DataFrame({"Purchase Count": pop_purchasec, "Item Price": pop_prices, "Total Purchase Value": pop_totpurchase})
pop_summary_df.head()

```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Sort the data frame
sort_summary_df = pop_summary_df.sort_values(by=["Total Purchase Value"], ascending = False)
sort_summary_df.head()

#Clean up Data Frame 
Sum_sort = sort_summary_df
Sum_sort["Item Price"] = Sum_sort["Item Price"].map("${:.2f}".format)
Sum_sort["Total Purchase Value"] = Sum_sort["Total Purchase Value"].map("${:.2f}".format)
Sum_sort.head()
```
